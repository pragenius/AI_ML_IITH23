```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True) #set a theme

x= np.random.random(100) #generate 100 random numbers
y=np.random.normal(size=100) #generate 100 numbers based on normal distribution

print (x[0:10:1]) #print 10 numbers from the top
```

```python
plt.hist(x) #display the histogram of x
```

```python
plt.hist(y) #plot hist distribution of y
```

```python
sns.histplot(y,bins=20) #display the histogram using seabon with bins =10
```

```python
sns.histplot(x,bins=20)
```

```python
sns.histplot(y,bins=20,kde=True) #curve and histogram plot
```

```python
sns.kdeplot(y,color='magenta',fill=True) #only kde plot
```

```python
sns.kdeplot(x,color='magenta',fill=True)
sns.kdeplot(y,color='cyan',fill=True) #two combined plots
```

```python
#subplots
```

```python
#subplots
plt.subplot(1,2,1)
sns.kdeplot(x,color='magenta',fill=True)
plt.subplot(1,2,2)
sns.kdeplot(y,color='cyan',fill=True)
```

```python
days=tips_df.day.unique()
day_data={}
for each_day in days:
  day_data[each_day]=tips_df[tips_df['day']==each_day]['day'].count()
print(day_data)
```

```python
sns.dataframe(day_data)
```

```python
tips_series=tips_df.groupby('day')['day'].count()
```

```python
sns.barplot(x=tips_series.index,y=tips_series.values)
```

```python
sns.relplot(data=tips_df,x='total_bill',y='tip') #relational plot
```

```python
sns.regplot(data=tips_df,x='total_bill',y='tip') #regression plot
```

```python
sns.lmplot(data=tips_df,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips_df[tips_df['sex']=='Male'],x='total_bill',y='tip')
```

```python
sns.pairplot(tips_df) #pairwise charting of all numerical data
```

```python
tips_series=tips_df.groupby('day')['day'].count()
plt.pie(tips_series,labels=tips_series.index,autopct='%0.2f%%',shadow=True,explode=[1,0,0,0])
```

```python
#donut chart
plt.pie(tips_series,labels=tips_series.index,autopct='%0.2f%%',shadow=True,wedgeprops=dict(width=0.25),explode=[0,0,1,0])
```
